#### Dropping unnecessary columns

```python
total_population_data=task[task['Indicator Code']=='SP.POP.TOTL']
total_population_sorted=total_population_data.sort_values(by="2020",ascending = False)
total_top_ten_countries=total_population_sorted.head(10)
print("top ten countries of total population \n")
print(total_top_ten_countries[['Country Code']])
```

## Bar plot

#### Top ten countries of total population in year 2020 and 2016

```python
plt.figure(figsize=(18,9))
plt.subplot(2,2,1)
sns.barplot(x="2020", y="Country Code", data=total_top_ten_countries, palette="coolwarm")
plt.title("Top Ten Countries of Total population(2020)",fontsize=10)
plt.xlabel("Total population", fontsize=20)
plt.ylabel("Country",fontsize=20)
plt.show()
# create a bar plot
plt.figure(figsize=(18,9))
plt.subplot(2,2,2)
sns.barplot(x="2016", y="Country Code", data=total_top_ten_countries, palette="coolwarm")
plt.title("Top Ten Countries of Total population(2016)",fontsize=10)
plt.xlabel("Total population",fontsize=20)
plt.ylabel("Country",fontsize=20)
plt.show()
```

#### Extarction of top 20 countries with highest male population

```python
male_population_data = task[task['Indicator Code'] == 'SP.POP.TOTL.MA.IN']
male_population_sorted = male_population_data.sort_values(by="2020",ascending=False)
male_top_twenty_countries = male_population_sorted.head()
print("Top Countries of male population")
print(male_top_twenty_countries[['Country Code']])
```

```python
#female population
female_population_data = task[task['Indicator Code'] == 'SP.POP.TOTL.FE.IN']
female_population_sorted = female_population_data.sort_values(by="2020",ascending=False)
female_top_twenty_countries = female_population_sorted.head()
print("Top Countries of female population")
print(female_top_twenty_countries[['Country Code']])
```

### Top 20 countries with highest male and female population in 2020

```python
plt.figure(figsize=(18,9))
plt.subplot(2,2,1)
sns.barplot(x="2020", y="Country Code", data=male_top_twenty_countries, palette="viridis")
plt.title("TOP Ten Countries of male population(2020)",fontsize=20)
plt.xlabel("male population", fontsize=20)
plt.ylabel("Country",fontsize=20)
plt.show()
# create a bar plot
plt.figure(figsize=(18,9))
plt.subplot(2,2,2)
sns.barplot(x="2016", y="Country Code", data=female_top_twenty_countries, palette="viridis")
plt.title("TOP Ten Countries of female population(2016)",fontsize=20)
plt.xlabel("female population",fontsize=20)
plt.ylabel("Country",fontsize=20)
plt.show()
```
